#RUST -> struct -> toml -> struct

```rust
:dep serde = { version = "1.0", features = ["derive"] }
:dep toml = "0.7.6"
```

```rust
use serde::Serialize;
use std::fs::File;
use std::io::Write;
use toml::to_string;
//use serde_derive::Deserialize;
use serde::Deserialize;
use std::collections::HashMap;
use toml;

//Main struct
#[derive(Debug, Deserialize, Serialize)]
struct Screener {
    url_for_screener: String,
    protocol: String, // #TODO! String or Enum
    //path: HashMap<String, Path>,
    //parameter: HashMap<String, Parameter>,
}
/*[parameter]
[screener.setting1]
field = "name"
value = "val"
timeout = 10
reload = yes # or no */

#[derive(Debug, Deserialize, Serialize)]
struct Parameter {
    field: String,
    value: String,
    timeout: u16,
    reload: String, // #TODO String enum
}
/*
[path]
  [path.step1]
  link_xpath = ""
  action = "click" # enum: click,dropdown,select
  action_value = ""
*/

#[derive(Debug, Deserialize, Serialize)]
struct Path {
    link_xpath: String,
    action: String, // #TODO String or enum
    action_value: String,
}

/*
[output]
[column1]
name = "NAME"
format = "FORMAT"
*/
#[derive(Debug, Deserialize, Serialize)]
struct Output {
    name: String,
    format: String, // #TODO String or enum
}

const TEXT_save: &str = r#"
url_for_screener = "https://finviz.com/"
protocol = "https"
timeout = 3600
"#;

const TEXT: &str = r#"
url_for_screener = hallo
protocol = "https"
timeout = 3600
[path]
  [path.step1]
  link_xpath = ""
  action = "click" # enum: click,dropdown,select
  action_value = ""
  [path.step2]
  link_xpath = ""
  action = "click" # enum click,dropdown,select
  action_value = ""
  
[parameter]
  [parameter.setting1]
  field = "name"
  value = "val"
  timeout = 10
  reload = yes # or no
  [screener.setting2]
  field = "name"
  value = "val"
  timeout = 10
  reload = yes # or no

[output]
    [column1]
    name = "NAME"
    format = "FORMAT"
"#;

fn main_tags()  {
    // from here
    // https://doc.rust-lang.org/rust-by-example/error/result.html
    let screener: Screener = toml::from_str(TEXT).unwrap();
    //let screener: Screener = toml::from_str("").unwrap();
    println!("{:#?}", screener);
    
}
```
